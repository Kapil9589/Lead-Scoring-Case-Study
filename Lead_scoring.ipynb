```python
# Now let's convert the bool(Yes / No)  into the numerical (1 and 0)
data['Do Not Email']=data['Do Not Email'].map({'Yes':1,'No':0})
data['A free copy of Mastering The Interview']=data['A free copy of Mastering The Interview'].map({'Yes':1,'No':0})
```

```python
data.info()
```

```python

# In null_data DataFrame the percentage of null values in the rows
# (Lead sores,TotalVisits,Page Views Per Visit,Last Activity) are allmost 1.5 % so let's drop the rows
column=['Lead Source','TotalVisits','Page Views Per Visit','Last Activity']

data = data.dropna(subset=column)
data.isnull().sum()
```

```python
data.info()
```

```python
data.head()
```

```python
# Treatment of outliers
num_var=data.select_dtypes(include=['int64','float64'])
num_var.head()
```

```python
cat_var=data.select_dtypes(include=['object'])
```

```python
cat_var.head()
```

```python
cat_var=pd.get_dummies(cat_var,drop_first=True,dtype=int)
```

```python
cat_var.head()
```

```python
data=pd.concat([num_var,cat_var],axis=1)
```

```python
data.head()
```

```python
data.info()
```

```python
data['TotalVisits']=data['TotalVisits'].astype(int)
data['Page Views Per Visit']=data['Page Views Per Visit'].astype(int)
```

```python
data.info()
```

```python
for co in data.columns:
  print(data[co].value_counts(normalize=True)*100)
```

```python
li=[]
for co in data.columns:
    max_percentage = data[co].value_counts(normalize=True).max() * 100
    if max_percentage <99:
        li.append(co)
```

```python
li
```

```python
# Remove the columns where feature have less importance
```

```python
data=pd.DataFrame(data,columns=li)
```

```python
data.shape
```

Scaling of data(fit data into scale)

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
data=pd.DataFrame(scaler.fit_transform(data),columns=data.columns)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
X=data.drop('Converted',axis=1)
y=data['Converted']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=100)
```

```python
print(X_train.shape ,X_test.shape ,y_train.shape ,y_test.shape)
```

```python
X_train=sm.add_constant(X_train)
```

```python
model = sm.GLM(y_train, X_train, family=sm.families.Binomial())
```

```python
result=model.fit()
```

```python
result.summary()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_train,result.predict(X_train).round()))
```

```python
print(accuracy_score(y_train,result.predict(X_train).round()))
```

```python
print(classification_report(y_train,result.predict(X_train).round()))
```

```python
# y_pred
X_test=sm.add_constant(X_test)
y_pred = result.predict(X_test)
```

```python
y_test = y_test.astype(int)
```

```python
y_test_pred_final = pd.DataFrame({'actual':y_test.values, 'predicted':y_pred})

y_test_pred_final.head()
```

```python
y_test_pred_final['predicted'] = y_test_pred_final.predicted.map(lambda x: 1 if x > 0.3 else 0)

# Let's see the head
y_test_pred_final.head()
```

```python
print(confusion_matrix(y_test,y_test_pred_final['predicted']))
print(accuracy_score(y_test,y_test_pred_final['predicted']))
print(classification_report(y_test,y_test_pred_final['predicted']))
```

```python
coefficients = result.params
print(coefficients.sort_values(ascending=False))
```

```python
# prompt: 1.	Which are the top three variables in your model which contribute most towards the probability of a lead getting converted? according this model

# Get the coefficients from the model
coefficients = result.params

# Sort the coefficients in descending order
sorted_coefficients = coefficients.sort_values(ascending=False)

# Get the top three variables
top_three_variables = sorted_coefficients.index[:3]

# Print the top three variables
print(top_three_variables)
```
